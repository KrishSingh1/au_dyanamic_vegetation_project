# Assigning Sites to their Bioclimatic Groups

```python
import rasterio
import rasterio.mask
import fiona
import matplotlib.pyplot as plt
import numpy as np
import os 
import pandas as pd
```

```python
with fiona.open("../DATASETS/AusPlots_Location/AusPlots_Geometries_20240508.shp", "r") as shapefile:
    shapes = [feature["geometry"] for feature in shapefile]
    shapes_name = [properties["properties"]['Name'] for properties in shapefile]
```

```python
column_names = ['Group_Code']
site_group_data = pd.DataFrame(columns = column_names, index = shapes_name)
site_group_data.head()
```

```python
file_path = '../DATASETS/Australian Bioclimatic Regions/Australian_Bioclimatic_Regions.tif'
variable_mean = []
with rasterio.open(file_path) as src:
    print(src)

    for i, shape in enumerate(shapes):

        shape = [shape] # shape has to be in a list 
        shape_name = shapes_name[i] # reference the name of  the site 
        #print(shape_name)

        out_image, out_transform = rasterio.mask.mask(src, shape, all_touched = True, crop=True, nodata = -999999) # perform masking
        out_image_filter = out_image[out_image != -999999]
        image_mean = np.nanmean(out_image_filter)       
        variable_mean.append(image_mean)
        
site_group_data['Group_Code'] = variable_mean
```

```python
temp = dict()

for i in range(1,7):
    temp[i] = [len(site_group_data[site_group_data['Group_Code'] == i])]

print(temp)
```
