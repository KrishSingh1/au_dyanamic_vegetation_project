```python
super_aggregated = pd.read_csv("../DATASETS/Australian Bioclimatic Regions/ausplots_bioclimatic_super_group_aggregated_counts.csv")
```

```python
super_aggregated
```

```python
super_aggregated = super_aggregated.sort_values(by = ['Group_Code', 'vegetation_type']) # Sort by alphabetical order by two columns
super_aggregated.loc[super_aggregated.index == 20, 'comments'] = pd.NA # this comment did not need to be made? Set to NA
super_aggregated
```

Where the comments are set as NAN, means it is a separate group.

```python
# Manual Interpretation of what super class the other sub classes should fall under
new_adjustments = {
    'Desert Heath.shrub'              : 'Desert Shrub',
    'Desert Sedge'                    : 'Desert Forb',
    'Desert Shrub.Mallee'             : 'Desert Tree.Palm',
    'Desert Tree.Mallee'              : 'Desert Tree.Palm',
    'Temp/Med Bryophyte'              : 'Temp/Med Tussock.grass',
    'Temp/Med Chenopod'               : 'Desert Chenopod',
    'Temp/Med Forb'                   : 'Temp/Med Tussock.grass',
    'Temp/Med Heath.shrub'            : 'Temp/Med Shrub', 
    'Temp/Med Rush'                   : 'Temp/Med Tussock.grass',
    'Temp/Med Sedge'                  : 'Temp/Med Tussock.grass',
    'Temp/Med Shrub.Mallee'           : 'Temp/Med Tree.Palm',
    'Temp/Med Tree.Mallee'            : 'Temp/Med Tree.Palm',
    'Temp/Med Tussock.grass'          : 'Temp/Med Tussock.grass',
    'Tropical/Savanna Chenopod'       : 'Desert Chenopod',
    'Tropical/Savanna Hummock.grass'  : 'Desert Hummock.grass',
    'Tropical/Savanna Sedge'          : 'Tropical/Savanna Tussock.grass',
    'Tropical/Savanna Shrub'          : 'Desert Shrub',
    'Tropical/Savanna Tree.Mallee'     : 'Tropical/Savanna Tree.Palm'
}

superclassification_definitions = {}
```

```python
print(pd.DataFrame([new_adjustments.keys(), new_adjustments.values()]).T)
```

```python
super_number = 1 
super_names = [] 
super_codes = []
for i, v in super_aggregated.iterrows():
    if v[['comments']].isnull().values == True:
        super_name = v['Group_Code'] + ' ' + v['vegetation_type']
        superclassification_definitions[super_name] = {super_number: super_name}
        super_names.append(super_name)
        super_codes.append(super_number)
        super_number += 1
```

```python
superclassification_definitions
```

```python
list(superclassification_definitions.values())
```

```python
for i, v in super_aggregated.iterrows():
    if v[['comments']].isnull().values == False:
        key_search = v['Group_Code'] + ' ' + v['vegetation_type']
        superclassification_definitions[key_search] = superclassification_definitions[new_adjustments[key_search]] 
```

```python
superclassification_definitions
```

```python
classification_matrix = pd.DataFrame.from_dict(data = superclassification_definitions, orient =  'index')
classification_matrix
```

```python
for i, v in super_aggregated.iterrows():
    super_name = v['Group_Code'] + ' ' + v['vegetation_type']
    super_code = int(list(superclassification_definitions[super_name].keys())[0])
    super_aggregated.loc[super_aggregated.index == i, 'super_code'] = super_code
    super_aggregated.loc[super_aggregated.index == i, 'super_group'] = superclassification_definitions[super_name][super_code]
```
