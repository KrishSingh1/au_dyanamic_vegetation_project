```python
soils_char_not_missing = soils_char[soils_char['soil.char.texture_grade'] != 'NC']
soils_char_not_missing = soils_char_not_missing[soils_char_not_missing['soil.char.texture_grade'].isna() == False]
```

```python
columns = ['CL', 'CLS', 'CS', 'HC', 'L', 'LC', 'LMC', 'LS', 'MC', 'MHC', 'S',
       'SCL', 'SL', 'ZCL', 'ZL']
empty_char = pd.DataFrame(columns = columns, index = np.unique(soils_char_not_missing['soil.char.site_location_name']))
```

```python
missing_sites = []

for i in np.unique(soils_char_not_missing['soil.char.site_location_name']):
    sub = soils_char_not_missing[soils_char_not_missing['soil.char.site_location_name'] == i]
    counts = sub['soil.char.texture_grade'].value_counts()
    for j in counts.index:
        empty_char.loc[empty_char.index == i, j] = sub['soil.char.texture_grade'].value_counts()[j]
```

```python
empty_char = empty_char.fillna(value = 0)
```

```python
groups = []
for i in empty_char.index:
    groups.append(i[:6])
```

```python
empty_char['group'] = groups
```

```python
empty_char
```

```python
grouped_char = empty_char.groupby('group').sum()
```

```python
grouped_char
```

```python
classifications = {
    'S': [0, 5],
    'LS': [5, 5],
    'CS': [5, 10],
    'SL': [10, 20],
    'L' : [25, 25],
    'ZL': [25, 25],
    'SCL' : [20, 30],
    'CL': [30, 35],
    'CLS' : [30, 35],
    'ZCL' : [30, 35],
    'LC' : [35, 40],
    'LMC' : [40, 45],
    'MC': [45, 55],
    'MHC': [50, 50],
    'HC': [50, 50]
}
classifications = pd.DataFrame(classifications)
```

```python
classifications
```

```python
cly_min = [] 
cly_max = [] 
for i,v in grouped_char.iterrows():
    clay_classes = v[v > 0].index.to_list()
    min_range = np.min(classifications.loc[0,clay_classes].values)
    max_range = np.max(classifications.loc[1,clay_classes].values)
    cly_min.append(min_range)
    cly_max.append(max_range)
```

```python
grouped_char['cly_min'] = cly_min
grouped_char['cly_max'] = cly_max
```

```python
grouped_char
```

```python
cly_min = [] 
cly_max = [] 

empty_char['group'] = 0
for i,v in empty_char.iterrows():
    clay_classes = v[v > 0].index.to_list()
    min_range = np.min(classifications.loc[0,clay_classes].values)
    max_range = np.max(classifications.loc[1,clay_classes].values)
    cly_min.append(min_range)
    cly_max.append(max_range)
    
empty_char['cly_min'] = cly_min
empty_char['cly_max'] = cly_max
```
