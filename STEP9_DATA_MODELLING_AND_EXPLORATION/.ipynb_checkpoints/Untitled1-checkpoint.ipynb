```python
## Lets try WAAPIL0003
site_location_name = 'WAAPIL0003'
file_name = f'{site_location_name}_1987_2022.csv'
tmin = pd.read_csv(f'../DATASETS/Climate_Gridded/tmin/{file_name}', usecols = ['tmin', 'time'], parse_dates = ['time']).copy() # tmin
tmax = pd.read_csv(f'../DATASETS/Climate_Gridded/tmax/{file_name}', usecols = ['tmax', 'time'], parse_dates = ['time']).copy() # tmax 
```

```python
climate_data = tmin.copy().merge(tmax, left_on = 'time', right_on = 'time')
climate_data = climate_data.sort_values('time')
```

```python
climate_data['tmean'] = (climate_data['tmin'] + climate_data['tmax'])/2
```
